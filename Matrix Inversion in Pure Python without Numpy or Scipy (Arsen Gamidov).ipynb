```python
#На вход первые две строчки это размерность матрицы, а затем сами элементы построчно
n, m = int(input()), int(input())
matrix = []
for i in range(n):
   matrix.append([float(input()) for _ in range(m)])

for i in range(n):
   for j in range(m):
      print(matrix[i][j], end=',')
   print()
```

```python
matrix #Выведем матрицу
```

```python
#Транспонируем матрицу в несколько шагов
#Cначала используем генератор списков со вложенными циклами for чтобы узнать размерность транспонированной матрицы, используя исходную.
trans_matrx = [[0 for j in range(len(matrix))] for i in range(len(matrix[0]))]
print(trans_matrx)
```

```python
#Внешний цикл предназначен для строк, а вложенный – для столбцов
for i in range(len(matrix)):
  for j in range(len(matrix[0])):
    trans_matrx[j][i]=matrix[i][j]
```

```python
print(trans_matrx) #Выводим транспонированную матрицу
```

```python
#Функция нахождения определителя матрицы (детерминант)
def deternminant(m):
  #для матрицы 3х3
    if len(m)==3: 
        return m[0][0]*m[1][1]*m[2][2]+m[0][1]*m[1][2]*m[2][0]+m[0][2]*m[1][0]*m[2][1]-m[2][0]*m[1][1]*m[0][2]-m[2][1]*m[1][2]*m[0][0]-m[2][2]*m[1][0]*m[0][1]
    det=0
    for c in range(len(m)): 
        det+=((-1)**c)*m[0][c]*deternminant(minor(m,0,c))
    return det
```

```python
deternminant(matrix) #Обратная матрица существует
```

```python
#Нахождение алгебраических дополнений
def TransMatrxAlgb(m):
    #Только для матрицы 3х3
    if len(m) == 3:
        return [[(m[2][2]*m[1][1])-(m[2][1]*m[1][2]), -1*((m[2][2]*m[1][0])-(m[2][0]*m[1][2])), (m[2][1]*m[1][0])-(m[2][0]*m[1][1])],
                [-1*((m[2][2]*m[0][1])-(m[2][1]*m[0][2])), (m[2][2]*m[0][0])-(m[2][0]*m[0][2]), -1*((m[2][1]*m[0][0])-(m[2][0]*m[0][1]))],
                [(m[1][2]*m[0][1])-(m[1][1]*m[0][2]), -1*((m[1][2]*m[0][0])-(m[1][0]*m[0][2])), (m[1][1]*m[0][0])-(m[1][0]*m[0][1])]]
```

```python
TransMatrxAlgb(trans_matrx)
```

```python
n = 1/deternminant(matrix)
```

```python
def getMatrixInverse(m):
  return [list(map(lambda x: round(x*n,1), z)) for z in m]
```

```python
Inverse_of_a_Matrix = getMatrixInverse(TransMatrxAlgb(trans_matrx))
```
