```python
class myDeque:
    
    def __init__(self, s: str):
        self.queue = [i for i in s]
        
    def insert(self, s: str):
        self.queue.append(s)
        
    def pop(self):
        return self.queue.pop(0)
    
    def is_empty(self):
        if len(self.queue) == 0: 
            return True
        else:
            return False
    
    def __eq__(self, other):
        if len(self.queue) == len(other.queue) and self.queue[0] == other.queue[0]:
            return True
        else:
            return False
        
    def __gt__(self, other):
        if len(self.queue) > len(other.queue):
            return True
        elif len(self.queue) == len(other.queue):
            if self.queue[0] < other.queue[0]:
                return True
            else:
                return False
        else:
            return False
        
    def __lt__(self, other):
        return other > self


# AA == AB
a = myDeque('AB')
b = myDeque('AB')
assert a == b, 'Equality test failed'

# AA > A
a = myDeque('AA')
b = myDeque('A')

assert a > b, '"Greater than" test 1 failed'

# BA > AB
a = myDeque('BA')
b = myDeque('AB')

assert b > a, '"Greater than" test 2 failed'

# AB < BA
a = myDeque('AB')
b = myDeque('BA')

assert (b > a) == False, '"Greater than" test 3 failed'

# ABB > BBB
a = myDeque('ABB')
b = myDeque('BBB')
assert a > b

# AA < BBB
a = myDeque('AA')
b = myDeque('BBB')
assert a < b
```

```python
def parse_deque(a, b):
    
    while not (a.is_empty() and b.is_empty()):
        if a > b:
            print(a.pop())
        else:
            print(b.pop())
            
a = myDeque('AA')
b = myDeque('BBB')

parse_deque(a, b)
```
